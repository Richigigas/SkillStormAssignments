### No Auth

```python
import requests
import os

base_url ='https://pokeapi.co'
poke_api = '/api/v2'
poke_endpoint = '/pokemon/infernape'

pokemon_url = base_url + poke_api + poke_endpoint
response = requests.get(pokemon_url)
poke_data = response.json()
print(poke_data["species"]["name"])
```

### API Key

```python
from dotenv import load_dotenv
import pandas as pd

load_dotenv()

base_url ='https://api.nytimes.com'
books_api = '/svc/books/v3'
hardcover_endpoint = '/lists/current/hardcover-fiction.json'

books_url = base_url + books_api + hardcover_endpoint


api_key = os.environ['nytimes_api_key']
params = {'api-key': api_key}

response = requests.get(books_url, params=params)
data = response.json()
books = data["results"]["books"]
df = pd.json_normalize(books)
df.head()
```

### OAuth

```python
import requests
from requests_oauthlib import OAuth1

noun_api_key = os.environ['nounoauth_key']
noun_secret = os.environ['nounoauth_secret']

auth = OAuth1(noun_api_key, noun_secret)
endpoint = "https://api.thenounproject.com/v2/icon/1"

response = requests.get(endpoint, auth=auth)
```
